```python
# f)

data.dropna(subset=['Ocjena'], inplace=True)
data
```

```python
# g)

def najbolji_ispit(row):
    ispit1 = pd.to_numeric(row['Ispit1'], errors='coerce')
    ispit_popravni = pd.to_numeric(row['Ispit1_popravni'], errors='coerce')

    if pd.isna(ispit1) and pd.isna(ispit_popravni):
        return np.nan

    if pd.notna(ispit1) and pd.isna(ispit_popravni):
        return ispit1

    if pd.isna(ispit1) and pd.notna(ispit_popravni):
        return ispit_popravni

    return max(ispit1, ispit_popravni)

data['Ispit1_final'] = data.apply(najbolji_ispit, axis=1)

data
```

```python
# h)

def najbolji_ispit(row):
    ispit2 = pd.to_numeric(row['Ispit2'], errors='coerce')
    ispit_popravni = pd.to_numeric(row['Ispit2_popravni'], errors='coerce')

    if pd.isna(ispit2) and pd.isna(ispit_popravni):
        return np.nan

    if pd.notna(ispit2) and pd.isna(ispit_popravni):
        return ispit2

    if pd.isna(ispit2) and pd.notna(ispit_popravni):
        return ispit_popravni

    return max(ispit2, ispit_popravni)

data['Ispit2_final'] = data.apply(najbolji_ispit, axis=1)

data
```

```python
# i)

data.drop(columns=["Ispit1", "Ispit2", "Ispit1_popravni", "Ispit2_popravni"], inplace=True)
data
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import scale
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# a)

data = pd.read_csv('/content/drive/MyDrive/VI-labovi/LV2/izvjestaj.csv')
data.head()
```

```python
# b)

data.replace("/", np.nan, inplace = True)
si = SimpleImputer(strategy = 'median')

data[data.columns[data.columns.get_loc('Ispit1')]] = si.fit_transform(data['Ispit1'].values.reshape(-1, 1))
data[data.columns[data.columns.get_loc('Ispit2')]] = si.fit_transform(data['Ispit2'].values.reshape(-1, 1))

data.head()
```

```python
# c)

si = SimpleImputer(strategy = 'mean')

data[data.columns[data.columns.get_loc('Ispit1_popravni')]] = si.fit_transform(data['Ispit1_popravni'].values.reshape(-1, 1))
data[data.columns[data.columns.get_loc('Ispit2_popravni')]] = si.fit_transform(data['Ispit2_popravni'].values.reshape(-1, 1))

data.head()
```

```python
# d)

data['Ispit1'] = scale(data['Ispit1'])
data['Ispit2'] = scale(data['Ispit2'])

data.head()
```

```python
# e)

mm = MinMaxScaler()
data['Ispit1_popravni'] = mm.fit_transform(data['Ispit1_popravni'].values.reshape(-1,1))
data['Ispit2_popravni'] = mm.fit_transform(data['Ispit2_popravni'].values.reshape(-1,1))

data.head()
```

```python
# f)

data.replace(np.nan, 0, inplace=True)

data.head()
```

```python
# g)

ocjene = data.pop('Ocjena')

data.head()
```

```python
# h)

np_ocjene = np.array(ocjene)
np_data  = data.to_numpy()

print(np_ocjene)
print(np_data)
```

```python
# i)

X = np_data
Y = np_ocjene

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
```
