```python
# d)

def fact(n):
  f = 1

  for x in range (1, n + 1):
    f = f * x

  return f

def choose(n, k):
  return int( fact(n) / (fact(k) * fact(n - k)))

n = int(input("Unesite broj n: "))
k = int(input("Unesite broj k: "))

print("Faktorijel n: ", fact(n))
print("Binomni koeficijent n nad k: ", choose(n, k))
```

```python
# a)
n = np.linspace(1, 100, 100)
x = np.sin(n)
y = np.cos(n)

plt.figure(figsize=(10, 5))
plt.plot(n, x, label='sin(n)', color='blue')
plt.plot(n, y, label='cos(n)', color='red')
plt.xlabel('n')
plt.ylabel('Vrijednost')
plt.title('Preklopljeni')
plt.legend()
plt.grid()
plt.show()

# b)
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

axs[0].plot(n, x, color='blue')
axs[0].set_title(' sin(n)')
axs[0].set_xlabel('n')
axs[0].set_ylabel('sin(n)')
axs[0].grid()

axs[1].plot(n, y, color='red')
axs[1].set_title('cos(n)')
axs[1].set_xlabel('n')
axs[1].set_ylabel('cos(n)')
axs[1].grid()

plt.show()
```
